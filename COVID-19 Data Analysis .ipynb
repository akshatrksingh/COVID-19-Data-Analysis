```python
df_conf.rename(columns={'Province/State':'state','Country/Region': 'country'}, inplace=True)
df_death.rename(columns={'Province/State':'state','Country/Region': 'country'}, inplace=True)
df_conf.drop(['Lat','Long'], axis = 1, inplace= True)
df_death.drop(['Lat','Long'], axis = 1, inplace= True)
```

```python
df_conf1 = df_conf.groupby('country').sum().reset_index()
df_death1 = df_death.groupby('country').sum().reset_index()
```

```python
df_conf1
```

```python
df_death1
```

```python
max_conf = df_conf1.sort_values(by=['6/5/21'],ascending = False).head(10)
max_conf
```

```python
max_conf_temp = max_conf.drop(max_conf.iloc[:,1:-1], axis=1)
max_conf_temp
```

```python
max_death = df_death1.sort_values(by=['6/5/21'],ascending = False).head(10)
max_death
```

```python
max_death_temp = max_death.drop(max_conf.iloc[:,1:-1], axis=1)
max_death_temp
```

```python
plt.style.use('classic')
fig, ax = plt.subplots()
sns.barplot(data = max_conf_temp, x = '6/5/21', y = 'country')
y = max_conf_temp['6/5/21']
for index, value in enumerate(y):
    plt.text(value, index, str(value))
ax.set_xlabel('Total Confirmed Cases',fontsize=15)
ax.set_ylabel('Top 10 Countries',fontsize=15)
plt.style.use('classic')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot(data = max_death_temp, x = '6/5/21', y = 'country')
y = max_death_temp['6/5/21']
for index, value in enumerate(y):
    plt.text(value, index, str(value))
ax.set_xlabel('Total Deaths',fontsize=15)
ax.set_ylabel('Top 10 Countries',fontsize=15)
plt.style.use('classic')
plt.show()
```

```python
dfCT = df_conf1.set_index('country').T
fig, ax = plt.subplots()
ax.set_xlabel('Date')
ax.set_ylabel('Confirmed Cases')
x = list(dfCT.index.values) 
for i in range(10):
    y = list(dfCT[max_conf_temp['country'].iat[i]])
    plt.plot(x, y, label = max_conf_temp['country'].iat[i])
plt.xticks((0,125,250,375,500))
plt.title('Confirmed Cases Trend')
plt.legend(bbox_to_anchor=(1.4,1))
plt.show()
```

```python
dfDT = df_death1.set_index('country').T
fig, ax = plt.subplots()
ax.set_xlabel('Date')
ax.set_ylabel('Deaths')
x = list(dfDT.index.values) 
for i in range(10):
    y = list(dfDT[max_death_temp['country'].iat[i]])
    plt.plot(x, y, label = max_death_temp['country'].iat[i])
plt.xticks((0,125,250,375,500))
plt.title('Death Trend')
plt.legend(bbox_to_anchor=(1.4,1))
plt.show()
```
